```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

image_path = "/content/drive/MyDrive/zahra/j2.jpg"
img = cv2.imread(image_path)


if img is None:
    print("  .     .")
else:
    grid_size = 6
    height, width, _ = img.shape
    step_x = width // grid_size
    step_y = height // grid_size

    annotated_img = img.copy()
    for i in range(0, width, step_x):
        for j in range(0, height, step_y):
            x_end = min(i + step_x, width)
            y_end = min(j + step_y, height)

            square = img[j:y_end, i:x_end]

            mean_color = square.mean(axis=(0, 1))
            mean_color_int = tuple(map(int, mean_color))
            cv2.rectangle(annotated_img, (i, j), (x_end, y_end), (0, 255, 0), 1)

            text = f"{mean_color_int}"
            text_x = i + step_x // 4
            text_y = j + step_y // 2
            cv2.putText(
                annotated_img,
                text,
                (text_x, text_y),
                cv2.FONT_HERSHEY_SIMPLEX,
                0.4,
                (255, 255, 255),
                1,
                cv2.LINE_AA
            )

    cv2_imshow(annotated_img)
```

```python
import cv2
import numpy as np
import os
from google.colab.patches import cv2_imshow
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score

directory_path = "/content/drive/MyDrive/zahra/"

colors = {
    "": np.array([0, 255, 0]),
    "": np.array([255, 0, 0])
}

true_labels = []
predicted_labels = []
misclassified_images = []

def process_image(image_path):
    img = cv2.imread(image_path)

    if img is None:
        print(f" {image_path}  .")
        return

    grid_size = 6
    height, width, _ = img.shape
    step_x = width // grid_size
    step_y = height // grid_size

    blue_count = 0
    green_count = 0

    for j in range(0, height, step_y):
        for i in range(0, width, step_x):
            x_end = min(i + step_x, width)
            y_end = min(j + step_y, height)

            square = img[j:y_end, i:x_end]

            mean_color = square.mean(axis=(0, 1))

            closest_color = None
            min_distance = float("inf")

            for color_name, color_value in colors.items():
                distance = np.linalg.norm(mean_color - color_value)
                if distance < min_distance:
                    min_distance = distance
                    closest_color = color_name

            if closest_color == "":
                blue_count += 1
            elif closest_color == "":
                green_count += 1

    if blue_count > green_count:
        predicted_label = ""
    elif green_count > blue_count:
        predicted_label = ""
    else:
        predicted_label = ""

    file_name = os.path.basename(image_path).lower()
    true_label = "" if "s" in file_name else ""

    true_labels.append(true_label)
    predicted_labels.append(predicted_label)

    if true_label != predicted_label:
        misclassified_images.append((image_path, img))

    print(f"{image_path}: : {true_label}, : {predicted_label}")

image_files = [f for f in os.listdir(directory_path) if f.endswith(('.jpg', '.jpeg', '.png'))]

for image_file in image_files:
    image_path = os.path.join(directory_path, image_file)
    process_image(image_path)

print("\n---   ---")
print(":", accuracy_score(true_labels, predicted_labels))
print("\n :")
print(confusion_matrix(true_labels, predicted_labels))
print("\n :")
print(classification_report(true_labels, predicted_labels))

print("\n---    ---")
for image_path, img in misclassified_images:
    print(f" {image_path}:     ")
    cv2_imshow(img)
```
